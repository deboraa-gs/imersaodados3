- Valores muito próximos de 0: variáveis pouco ou não correlacionadas

```python
corr_gene = dados.loc[:,'g-0':'g-50']
corr_celular = dados.loc[:,'c-0':'c-50']

geneCelula = pd.concat([corr_gene, corr_celular], axis=1).corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(geneCelula, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(25, 15))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(geneCelula, mask=mask, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .8})
```

**Analise de causalidade:**

* 0.9 a 1 positivo ou negativo indica uma correlação muito forte.

* 0.7 a 0.9 positivo ou negativo indica uma correlação forte.

* 0.5 a 0.7 positivo ou negativo indica uma correlação moderada.

* 0.3 a 0.5 positivo ou negativo indica uma correlação fraca.

* 0 a 0.3 positivo ou negativo indica uma correlação desprezível.

```python
# Gere uma máscara para o triângulo superior
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

* mask: para cobrir parte do mapa de calor. Pega um array booleano ou um dataframe como argumento; quando definidas, as células tornam-se invisíveis para valores onde a máscara é verdadeira

* np.triu(): função do numpy para isolar o triângulo superior de uma matriz enquanto transforma todos os valores no triângulo inferior em 0.
np.tril () faria o mesmo, apenas para o triângulo inferior.

* np.ones_like(): muda todos os valores isolados para 1.

* cmap: define mapa de cores específico.

* vmin, vmax: define o intervalo de valores, base para o mapa de cores

* center: flutuação para centralizar o mapa de cores; se nenhum cmap for especificado, mudará as cores no mapa de cores padrão; se definido como 'true' muda todas as cores do mapa de cores para azul

* cbar: quando definido como False, a barra de cores (que serve como uma legenda) desaparece

* annot - quando definido como True, os valores de correlação se tornam visíveis nas células coloridas

OBS.: é preciso separar a base de dados em base de treino e base de teste.

Treino: onde o modelo vai aprender.

Teste: verificar o aprendizado do modelo.

Se treinar e testar o modelo com o mesmo conjunto de dados é mais fácil dele prever corretamente. A base de dados de teste, funciona como um novo conjunto onde pode-se medir o quanto o modelo está acertando, dado que já se tem a variável resposta real.

A divisão é feita usando a função do Scikit-Learn chamada train_teste_split

A proporção entre treino e teste, varia de acordo com o volume de dados. Mas, usualmente, encontramos proporções 30/70, 25/75 ou 20/80. Aqui, foi definido como 0,2 ou 20%, isso significa que a base de treino será composta pelos 80% restante.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split #instalar se não estiver no colab!

x = dados_combinados.select_dtypes('float64') #todos os g's e c's
y = dados_combinados['ativo_moa'] #dados da coluna especificada
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376) 
#test_size: tamanho definido para a base de dados
#stratify: recebe o nome da coluna que deve levar em consideração a proporção de classes 
#ramdom_state: para a divisão ser sempre a mesma. P/ a % não mudar

modelo_rlogistica = LogisticRegression(max_iter=1000) #max_iter = 1000 para convergir para um resultado
modelo_rlogistica.fit(x_treino, y_treino) #.fit para fazer o treinamento do modelo
modelo_rlogistica.score(x_teste, y_teste) #.score() para calcular a acurácia. Quantas predições o modelo acertou na base de teste.
```

* O modelo está acertando ~62% das previsões!

```python
#usando o DummyClassifier do Scikit-Learn como base comparativa 
from sklearn.dummy import DummyClassifier

#accuracy_score: receberá como parâmetros a base teste da variável resposta (y_teste) e as previsões do modelo.
from sklearn.metrics import accuracy_score 

modelo_dummy = DummyClassifier('most_frequent') #estratégia de classificação
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste) #.predict para calcular as previsões
accuracy_score(y_teste, previsao_dummy)
```

```python
#treinando um modelo de arvore de decisão
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

Estratégia de classificação: [Link documentação de DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

```python
#plotando figura da árvore de decisão
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

A figura é dividida em 4 níveis.
* 1º nivel: composto por apenas um quadro (nó raíz ou raíz).São os melhores atributos que podem dividir a base de dados em dois conjuntos distintos.

* Entre o primeiro e o segundo nível (e nos demais níveis), temos duas setas que são chamadas de ramos e subdividem o nível em dois outros conjuntos que separam novamente os dados em mecanismo de ação ativado ou não ativado, a partir de uma decisão.

* O último nível da árvore (mais abaixo), apresenta n quadros que apresentam características que devem ser levadas em conta na hora da classificação e se dividem em class = ativado e class = não ativado. Esse nível não apresenta mais ramificações e, esses quadros são chamados de folhas.

OBS.: max_depth representa a profundidade da árvore.

[Link sobre árvore de decisão](https://minerandodados.com.br/arvores-de-decisao-conceitos-e-aplicacoes/)

```python
#aumentando os níveis da árvore de decisão e adicionando um 'for' para repetição dentro de um range definido

from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

teste=[] #lista com resultados da acurácia nos dados de teste 
treino=[] #lista com resultados da acurácia nos dados de treino
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i) #variando a profundidade da árvore
    modelo_arvore.fit(x_treino, y_treino) # ajuste do modelo a partir das bases de treino
    teste.append(modelo_arvore.score(x_teste, y_teste)) #armazena o resultado na lista de teste
    treino.append(modelo_arvore.score(x_treino, y_treino)) #armazena o resultado na lista de treino
```
