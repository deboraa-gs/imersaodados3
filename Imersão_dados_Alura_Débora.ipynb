```python
mapa = {"droga":"composto"}
dados.rename(columns = mapa, inplace = True) #muda o nome da coluna direto no dataframe
dados.head()
```

Atenção! Tudo o que foi feito com 'droga' antes não funcionará mais. 

```python
#veerificando os compostos com maior ocorrencia. No intervalo de zero a cinco
compostos = dados['composto'].value_counts()[0:5]
compostos
```

```python
#para obter somente o codigo dos 5 compostos mais utilizados
codigo_compostos = dados['composto'].value_counts().index[0:5]
codigo_compostos
```

Usando a função query do Pandas para filtrar a base de dados

Selecionando apenas as linhas nas quais o composto esteja dentro da lista cod_composto (lista que representa os 5 compostos mais testados no experimento)

```python
#aplicando filtros na coluna 'composto' com o query
dados.query('composto in @codigo_compostos')
```

'composto' é a coluna a ser verificada no dataframe e 'codigo_compostos' é a lista com os top 5 compostos

O @ é necessário para informar o query que 'codigo_composto' é uma variável que já foi definida fora da função.

```python
#Plotando gráficos usando seaborn com filtro aplicado na coluna 'composto'

import seaborn as sns
import matplotlib.pyplot as plt

sns.set() #usado para ficar no padrão do seaborn

plt.figure(figsize=(8,6)) #dimensões do gráfico
ax = sns.countplot(x = 'composto', data = dados.query('composto in @codigo_compostos'))
ax.set_title('Top 5 compostos')
plt.show()
```

Analises do gráfico Boxplot:
- Caixa no meio é a mediana (linha no meio da caixa que é o ponto onde metade dos dados estão na direita e a outra metade para a esquerda)
- Outliers (pontos acima ou abaixo do eixo principal do gráfico que representam valores discrepantes para mais ou para menos)
- Maior concentração dos dados (caixa principal que representa onde está a maior parte dos dados - primeiro quartil (25%) e terceiro quartil (75%)) 
- Máximos e mínimos desconsiderando os outliers (linhas laterais à caixa principal).

```python
ax = sns.countplot(x ='composto', data = dados.query('composto in @codigo_compostos'), palette = "mako", order = codigo_compostos)
ax.set_title('Top 5 Compostos', fontsize=16, fontweight='bold')
ax.set_ylabel('Quantidade', fontsize = 15)
ax.set_xlabel('Composto', fontsize = 15)
sns.set_style(style = "whitegrid")

plt.show()

#Documentação seaborn para definir a palette: https://seaborn.pydata.org/tutorial/color_palettes.html
```

Estilos de gride do seaborn:
- Darkgrid, whitegrid, dark, white, e ticks. 
- Padrão: darkgrid. 
- Cinza branco ajuda a evitar que a grade concorra com as linhas que representam os dados. O whitegrid tema é semelhante, mas é mais adequado para gráficos com elementos de dados pesados.
